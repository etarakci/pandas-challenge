```python
# Pandas HW for Data Analytics 24 week Bootcamp
# by Erendiz Tarakci
# last updated 3/38/2020
# observable trends listed above

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data.SN.value_counts())
print("Total Players: " + str(total_players))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item Name"].value_counts())

no_purchases = len(purchase_data)

total_rev = purchase_data.Price.sum()

avg_price = round(total_rev/no_purchases,2)

data = [[unique_items,"$"+ str(avg_price),no_purchases,"$" + str(total_rev)]]
purchasing_analysis_df = pd.DataFrame(data, columns = ['Number of Unique Items', 'Average Price','Number of Purchases','Total Revenue'])
purchasing_analysis_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#get genders of unique players

gender_group = purchase_data.groupby("Gender")
player_gender = pd.DataFrame(gender_group["SN"])
player_gender.loc[0,1] = player_gender.loc[0,1].unique()
player_gender.loc[1,1] = player_gender.loc[1,1].unique()
player_gender.loc[2,1] = player_gender.loc[2,1].unique()

no_female = len(player_gender.loc[0,1])
no_male = len(player_gender.loc[1,1])
no_other = len(player_gender.loc[2,1])

total_gender = no_female + no_male + no_other

percent_female = round(no_female/total_gender*100,2)
percent_male = round(no_male/total_gender*100,2)
percent_other = round(no_other/total_gender*100,2)

data = [["Male", no_male, str(percent_male) + "%"],["Female", no_female, str(percent_female) + "%"],["Other/Non-Disclosed",no_other,str(percent_other) + "%"]]
gender_demographics = pd.DataFrame(data,columns=["Gender","Total Count","Percentage of Players"])
gender_demographics = gender_demographics.set_index("Gender")
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count aka no purcahses per gender
gender_price = pd.DataFrame(gender_group["Price"])

purchase_count_female = len(gender_price.loc[0,1])
purchase_count_male = len(gender_price.loc[1,1])
purchase_count_other = len(gender_price.loc[2,1])

#total purchase value
total_pvalue_female = gender_price.loc[0,1].sum()
total_pvalue_male = gender_price.loc[1,1].sum()
total_pvalue_other = gender_price.loc[2,1].sum()

#avg purchase price per gender
avg_female = round(total_pvalue_female/purchase_count_female,2)
avg_male = round(total_pvalue_male/purchase_count_male,2)
avg_other = round(total_pvalue_other/purchase_count_other,2)

#avg purchase total per person
avg_per_person_female = round(total_pvalue_female/no_female,2)
avg_per_person_male = round(total_pvalue_male/no_male,2)
avg_per_person_other = round(total_pvalue_other/no_other,2)

data = [["Female",purchase_count_female,"$" + str(avg_female),"$" + str(total_pvalue_female),"$" + str(avg_per_person_female)],["Male",purchase_count_male,"$" + str(avg_male),"$" + str(total_pvalue_male),"$" + str(avg_per_person_male)],["Other/Non-Disclosed",purchase_count_other,"$" + str(avg_other),"$" + str(total_pvalue_other),"$" + str(avg_per_person_other)]]
purchasing_analysis_gender = pd.DataFrame(data,columns = ["Gender","Purchase Count","Average Purcahse Price","Total Purchase Value","Avg Total Purchase per Person"])
purchasing_analysis_gender = purchasing_analysis_gender.set_index("Gender")
purchasing_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99,34.99, 39.99, 100]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

age_group = purchase_data.groupby("Age Group")
age_count = age_group["SN"].nunique()
age_percent = age_count/total_players*100

age_demographics = pd.DataFrame(age_count)
age_demographics["Percentage of Players"] = round(age_demographics["SN"]/total_players*100,2)
age_demographics.rename(columns = {"SN":"Total Count"}, inplace=True)

#formatting
age_demographics["Percentage of Players"] = [str(p) + "%" for p in age_demographics["Percentage of Players"]]
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase_data binned in previous section
purchase_analysis_age = pd.DataFrame(age_group["SN"])

purchase_analysis_age.rename(columns={0:"Age Group",1:"Purchase Count"},inplace=True)
purchase_analysis_age["Purchase Count"] = [len(x) for x in purchase_analysis_age["Purchase Count"]]    

age_purchase_total = pd.DataFrame(age_group["Price"])
age_purchase_total.rename(columns={0:"Age Group",1:"Total Purchase Value"},inplace=True)
age_purchase_total["Total Purchase Value"] = [x.sum() for x in age_purchase_total["Total Purchase Value"]]

purchase_analysis_age = pd.merge(purchase_analysis_age,age_purchase_total, on="Age Group")
purchase_analysis_age["Average Purchase Price"] = round(purchase_analysis_age["Total Purchase Value"] / purchase_analysis_age["Purchase Count"],2)
purchase_analysis_age["Avg Total Purchase per Person"] = round(purchase_analysis_age["Total Purchase Value"] / age_demographics["Total Count"].tolist(),2)
purchase_analysis_age = purchase_analysis_age[["Age Group","Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]
purchase_analysis_age.set_index("Age Group",inplace=True)

#formatting
purchase_analysis_age["Average Purchase Price"] = ["$" + str(p) for p in purchase_analysis_age["Average Purchase Price"]]
purchase_analysis_age["Total Purchase Value"] = ["$" + str(round(p,2)) for p in purchase_analysis_age["Total Purchase Value"]]
purchase_analysis_age["Avg Total Purchase per Person"] = ["$" + str(p) for p in purchase_analysis_age["Avg Total Purchase per Person"]]

purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Get sum purchases and total purchase count for each SN
sn_group = purchase_data.groupby("SN")
sn_price = pd.DataFrame(sn_group["Price"])
sn_price.rename(columns={0:"SN",1:"Purchased"},inplace=True)
sn_price.set_index("SN",inplace=True)

sn_price["Purchase Count"] = [len(x) for x in sn_price["Purchased"]]
sn_price["Total Purchase Value"] = [x.sum() for x in sn_price["Purchased"]]
sn_price["Average Purchase Price"] = round(sn_price["Total Purchase Value"]/sn_price["Purchase Count"],2)
top_spenders = sn_price[["Purchase Count","Average Purchase Price","Total Purchase Value"]]

top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

top_spenders["Average Purchase Price"] = ["$" + str(p) for p in top_spenders["Average Purchase Price"]]
top_spenders["Total Purchase Value"] = ["$" + str(round(p,2)) for p in top_spenders["Total Purchase Value"]]

top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular_items = purchase_data[["Item ID","Item Name","Price"]]
item_group = most_popular_items.groupby(["Item ID","Item Name"]).agg({"Price":[len,sum]})
item_group.columns = ["Purchase Count","Total Purchase Value"]
item_group_df = pd.DataFrame(item_group) 
item_group_df["Item Price"] = round(item_group_df["Total Purchase Value"]/item_group_df["Purchase Count"],2)
item_group_df = item_group_df.sort_values("Purchase Count", ascending=False)

most_popular_items = item_group_df[["Purchase Count", "Item Price","Total Purchase Value"]]

#Formatting
formatted_most_popular_items = most_popular_items.copy()
formatted_most_popular_items["Item Price"] = ["$" + str(p) for p in most_popular_items["Item Price"]]
formatted_most_popular_items["Total Purchase Value"] = ["$" + str(round(p,2)) for p in most_popular_items["Total Purchase Value"]]
formatted_most_popular_items["Purchase Count"] = [round(p) for p in most_popular_items["Purchase Count"]]

formatted_most_popular_items.head()

```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable_items = most_popular_items.sort_values("Total Purchase Value", ascending=False)

#Formatting
formatted_most_profitable_items = most_profitable_items.copy()
formatted_most_profitable_items["Item Price"] = ["$" + str(p) for p in formatted_most_profitable_items["Item Price"]]
formatted_most_profitable_items["Total Purchase Value"] = ["$" + str(round(p,2)) for p in formatted_most_profitable_items["Total Purchase Value"]]
formatted_most_profitable_items["Purchase Count"] = [round(p) for p in formatted_most_profitable_items["Purchase Count"]]

formatted_most_profitable_items.head()

formatted_most_profitable_items.head()
```
